```python
# when call it 5 times ,the output =5
def counter():
    count = 0
    def inner():
        nonlocal count
        count += 1
        return count
    return inner

count_function = counter()

#with generator
def counter_gen():
    count = 0
    while True:
        count += 1
        yield count


```
