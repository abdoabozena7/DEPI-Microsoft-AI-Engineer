```python

class Car:
    def __init__(self, model, color, speed=0):
        self.model = model
        self.color = color
        self.speed = speed

    def drive(self, increase_speed):
        self.speed += increase_speed
        return f"The {self.color} {self.model} is now driving at {self.speed} km/h."
```
