```python

class Human:
    def __init__(self, name, age, height):
        self.name = name
        self.age = age
        self.height = height

    def walk(self):
        return f"{self.name} is walking."

    def talk(self):
        return f"{self.name} is talking."
```

```python
ali = Human("ali", 22, 150)
lila = Human("lila", 30, 120)
```

```python
class Employ:
    def __init__(self, name, salary):
        self.name = name
        self._salary = salary  

    @property
    def salary(self):
        return self._salary

    @salary.setter
    def salary(self, value):
        if value < 0:
            raise ValueError("Salary cannot be negative.")
        self._salary = value
```
