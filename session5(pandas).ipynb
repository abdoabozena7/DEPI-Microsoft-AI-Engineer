```python
import pandas as pd
```

```python
df = pd.read_csv('D:\\projects\\DEPI-Microsoft-AI-Engineer\\datasets\\employees.csv')

```

```python
df.head()
```

```python
salaryByAge = df.groupby('Age')['Salary'].mean()
print("Average Salary by Age:")
print(salaryByAge)
```

```python
df.describe()
```

```python
# salary > 4000 , age < 30
mask = (df['Salary'] > 4000) & (df['Age']<30)

df[mask]
```

```python
mask = (df['Department']=="IT") & (df['Salary']>60000)
df[mask]
```

```python
df['Bonus'] = df['Salary']*.1
df.head()
```

```python
# Filter employees with salary > 5000.
salary_mask = df['Salary'] > 5000

# Add new column "Tax" = 15% of salary.
df['Tax'] = df['Salary'] * 0.15

# Increase salary by 1000 for filtered employees.
df.loc[salary_mask, 'Salary'] = df.loc[salary_mask, 'Salary'] + 1000

# Show only Name + Tax columns for filtered employees.
df.loc[salary_mask, ['Name', 'Tax']]
```

```python

avg_age_by_department = df.groupby('Department')['Age'].mean()
print("Average Age by Department:")
print(avg_age_by_department)

avg_bonus_by_age_group = df.groupby('Age')['Bonus'].mean()
print("\nAverage Bonus by Age Group:")
print(avg_bonus_by_age_group)

df['Joining_Date'] = pd.to_datetime(df['Joining_Date'])
hires_by_year = df.groupby(df['Joining_Date'].dt.year)['Name'].count()
print("\nHires by Year:")
print(hires_by_year)
```

```python
df['Quarter'] = df['Joining_Date'].dt.to_period('Q')
hires_by_quarter = df.groupby('Quarter')['Name'].count()
print("\nHires by Quarter:")
print(hires_by_quarter)
```

```python
# based on department
# min salary / avg bouns / avg age
avg_stats_by_department = df.groupby('Department').agg({
    'Salary': 'min',
    'Bonus': 'mean',
    'Age': 'mean'
})
print("\nAverage Stats by Department:")
print(avg_stats_by_department)

```

```python
age=df.groupby('Age')
salary = age['Salary']
print("Average Salary by Age:")
print(salary.mean())
```

```python
df['Joining_Date'] = pd.to_datetime(df['Joining_Date'])
HiresByYear = df.groupby(df['Joining_Date'].dt.year)['Name'].count()
print("Hires by Year:")
print(HiresByYear)



dates = df['Joining_Date']
years = dates.dt.year
g = df.groupby(years)
hires_by_year = g['Name'].count()
print("Hires by Year:")
print(hires_by_year)
```

```python
# pivot table
# max slary per dep
# count emp per dep

pivot = df.pivot_table(index='Department', values='Salary', aggfunc='max')
print("Max Salary per Department:")
print(pivot)
```

```python
#مجموع الحوافز اللى طلع لكل قسم على مدار اجزاء السنه 

pd.pivot_table(
    df,
    values='Salary',
    index='Department',
    columns='Quarter',
    aggfunc='mean',
    fill_value=0)
```
